- self-balancing binary search tree (BST)
- diff btw heights of LEFT and RIGHT subtree = balance factor
- balance factor =< 1

- if more than 1, tree needs rotations
- ensures the tree is nearly balanced, o(logn) insertion, delete, lookup
- Left rotation, right rotation, left-right rotation, right-left rotation
- node is balanced if balance factor = -1, 0, 1

advantages
- ensures o(logn)
- balanced structure, height of tree is always logarithmic to the nodes
- regular BST can be linked lsit = o(n) 

efficient dynamic operations
- dynamic datasets: insertions and deletions are frequent

predictable performance
- regular BSTS performance depends on the order of insertions,
- AVLs can be predictable

balanced structure 
- provides shallow tree: minimizes comparisions
- efficient range queries: all values are sorted

memory efficient
- o(n) + height and balance factor values (small)

applications:
- databse indexing
- language libraries: maps sets
- file systems
- networking: routing tables
