- BFS: level order insertion
- DFS: traversal, preorder, inorder, postorder

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


class BinaryTree:
    def __init__(self):
        self.root = None
    
    def insert(self, value):
        new_node = Node(value)

        # check tree is empty
        if self.root == None:
            self.root = new_node
            return
        
        # level order traversal to find empty place
        queue = [self.root]

        while queue:
            current = queue.pop(0)
            if current.left == None:
                current.left = new_node
                return
            else:
                queue.append(current.left)


            
            if current.right == None:
                current.right = new_node
                return
            else:
                queue.append(current.right)



    def preorder(self, node):
        if node == None:
            return
        
        print(node.value, end= " ")
        self.preorder(node.left)
        self.preorder(node.right)


    def inorder(self, node):
        if node == None:
            return
        
        self.inorder(node.left)
        print(node.value, end= " ")
        self.inorder(node.right)

    def postorder(self, node):
        if node == None:
            return
        
        self.inorder(node.left)
        self.inorder(node.right)
        print(node.value, end= " ")

    def levelorder(self):
        if self.root == None:
            return
        

        # create queue for checked items and its children
        queue = [self.root]
        while queue:
            current = queue.pop(0)
            print(current.value, end= " ")
            if current.left:
                queue.append(current.left)

            if current.right:
                queue.append(current.right)
```

```python
tree = BinaryTree()
tree.insert(1)
tree.insert(2)
tree.insert(3)
tree.insert(4)
tree.insert(5)
tree.insert(6)
tree.insert(7)

print("Pre-order Traversal:")
tree.preorder(tree.root)  #  1 2 4 5 3 6 7

print("\nIn-order Traversal:")
tree.inorder(tree.root)  #  4 2 5 1 6 3 7

print("\nPost-order Traversal:")
tree.postorder(tree.root)  # 4 5 2 6 7 3 1

print("\nLevel-order Traversal:")
tree.levelorder()  # 1 2 3 4 5 6 7
```
